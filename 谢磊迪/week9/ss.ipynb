```python
class MyClass:
    def __init__(self):
        self.attribute1 = []
        self.attribute2 = []

    def use_attribute(self, attr_name):
        # 使用 getattr 动态获取属性
        attribute_value = getattr(self, attr_name)
        attribute_value.append(1)
        
        # print(attribute_value)
    def aa(self,attr_name):
        for i in range(11):
            self.use_attribute(attr_name)


# 实例化对象
obj = MyClass()

# 根据传入的属性名动态访问属性
obj.aa('attribute1')  # 输出: This is attribute 1
obj.use_attribute('attribute2')  # 输出: This is attribute 2
```
